### Loading our data

First we'll load up the data.

```python
import pandas as pd
selected_df = pd.read_csv('./olist_2017_2018.csv')
```

```python
target_cols = 'price'

id_cols = ['order_id', 'customer_id', 'product_id']
geo_cols = ['customer_zip_code_prefix']
customer_cols = ['customer_city', 'customer_state']
product_cols = ['product_category_name']
purchase_time_cols = ['purchase_timeYear', 'purchase_timeMonth', 'purchase_timeWeek',
       'purchase_timeDay', 'purchase_timeDayofweek', 'purchase_timeDayofyear',
       'purchase_timeElapsed']
```

### Exploring the data

Below, with the 80 - 20 principle in mind, explore different dimensions of the data.  And dig deeper where necessary.

* Customer columns


Below we first explore the various customer columns.  And we can do so just by passing them into our `build_and_print` function.

```python
from item_lib import *

target = 'price'
agg_name = 'cum pct total sales'

geo_results = build_and_print(selected_df, customer_cols,
                              target, agg = 'sum',
                              agg_name = agg_name, y_col = -1, y_range=[0, .9])
```

```python
geo_results['customer_city'][:4]
```

So above, we can see that Sao Paulo tends to dominate -- it has twice as many sales as Rio.  And the top four cities account 25% of all of our customers.

* Product columns

We can perform similar analysis on the product columns.

```python
results = build_and_print(selected_df, product_cols,
                          target, agg = 'sum', y_col = -1,
                          agg_name = 'cum_pct_total_sales', limit = 8)
```

So again we can see something similar.

```python
results.keys()
```

```python
results['product_category_name'][:5]
```

The top five categories account for 41 percent.  And the top ten account for 60%.

```python
indexed_cumsum = results['product_category_name'].reset_index()
```

```python
import plotly.graph_objects as go
scatter = go.Scatter(x = indexed_cumsum.index/73, y = indexed_cumsum['cum_pct'])
go.Figure(scatter, layout = {'title': 'Product pct sales vs pct products'})
```

* Zipcode columns

```python
geo_cols = ['customer_zip_code_prefix']

results = build_and_print(selected_df, geo_cols,
                          target, agg = 'sum', y_col = -1,
                          agg_name = 'pct_total_sales', y_range = [0, .1])
```

Ok, so even with something as diverse as zipcodes, we can see some concentration of the data.

```python
cum_zipcode_df = results['customer_zip_code_prefix'].reset_index()
cum_zipcode_df[:10]
```

```python
len(cum_zipcode_df.index)
```

```python
import plotly.graph_objects as go
scatter = go.Scatter(x = cum_zipcode_df.index/len(cum_zipcode_df.index), y = cum_zipcode_df.cum_pct)
go.Figure(scatter)
```

So we can see that our *top ten* zipcodes (out of 14,000) account for 15% of total sales.  And the top 13% of zipcodes account for 50% of sales.

#### Digging deeper 

Ok, so we identified a couple of dimensions across where we can look at our data:

1. Product categories (the top five account for 41% of sales)
2. Regions (Top 4 account for 25% of sales, with Rio accounting for 15%)
3. Zipcodes - Less than .1% of zipcodes accounted for 15% of sales

Let's choose `non_beleza_saude` products, as a comparison.

```python
non_beleza_saude_df = selected_df[selected_df['product_category_name'] != 'beleza_saude']
```

After selecting this subset of the data, we can then explore it to see how this subset again.  Below, we copy over our column groups again (just so we can look at them).

```python
target_cols = 'price'

id_cols = ['order_id', 'customer_id', 'product_id']
geo_cols = ['customer_zip_code_prefix']
customer_cols = ['customer_city', 'customer_state']
product_cols = ['product_category_name']
purchase_time_cols = ['purchase_timeYear', 'purchase_timeMonth', 'purchase_timeWeek',
       'purchase_timeDay', 'purchase_timeDayofweek', 'purchase_timeDayofyear',
       'purchase_timeElapsed']
```

And then we can explore how this breaks down.

```python
customer_results_beleza = build_and_print(beleza_saude_df, customer_cols, target, agg = 'sum', 
                          agg_name = 'pct_total_sales_beleza', y_col = -1)
```

Looking at the above, the customer results look similar to our overall data.

And what if we compare to zipcodes of this product, to sales by zipcode overall.

```python
beleza_zipcode_results = build_grouped_by(beleza_saude_df, geo_cols, target,
                                          agg = 'sum',
                                          agg_name = None, pct = True)
```

```python
beleza_zipcode_results[:3]
```

```python
non_beleza_saude_zip = build_grouped_by(non_beleza_saude_df, geo_cols, target,
                                          agg = 'sum',
                                          agg_name = None, pct = True)


non_beleza_saude_zip[:3]
```

It does look like that our beleza products are more concentrated in certain zipcodes than products overall.  Let's plot the cumulative percentage with plotly.

```python

import plotly.graph_objects as go

indexed_df = beleza_zipcode_results.reset_index()
scatter = go.Scatter(x = indexed_df.index/len(indexed_df.index),
                     y = indexed_df['cum_pct'])
go.Figure(data = [scatter], layout = {'title': 'Percent spend vs Percent zipcodes'})
```

We can see that five percent of zipcodes account for 25% of sales of beleza.  From here, plotting the data geographically might be beneficial -- to look for similarities in our top regions.

So how does this help us?

* In the short term, this can help us with inventory managment -- making sure that we deliver products to this market.

* Over the longer term, we could look to see marketing opportunities with insight.  Can we market to the product to similar zipcodes?  Is there a reason why sales of the product lag in other (similar) zipcodes?  This may involve both qualitative and quantitative info.
