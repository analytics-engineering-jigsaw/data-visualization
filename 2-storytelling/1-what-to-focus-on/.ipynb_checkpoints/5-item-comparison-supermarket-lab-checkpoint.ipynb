### Loading our Data

We can begin by loading our data which will come from the [Supermarket sales data](https://www.kaggle.com/datasets/aungpyaeap/supermarket-sales).

```python
import pandas as pd
url = "https://raw.githubusercontent.com/analytics-engineering-jigsaw/data-visualization/main/2-storytelling/1-what-to-focus-on/supermarket_sales.csv"
supermarket_df = pd.read_csv(url)
```

Now to explore the dataset, first slice the first few rows of data.

```python
supermarket_df[:3]
```

#### Getting an overview - dimensions, and target

Now that we've loaded the data, the next step is to get an overview.  

1. Find the dimensions

Our first step is to ensure we have the dimensions of the data.  By dimensions of the data we mean -- what does each row represent.  Our first clue is to look at the invoice-id, but looking at the product line, price and quantify -- it looks like each row can only consist of a customer buying a single product at various quantities. 

> So each row is a purchase of a single product by a customer.

2. Determine the target 

Our second step is to determine a target.  

There are actually a few candidates here: 
* Price
* Quantity
* Total
* Gross income
* Gross margin percentage

We can look at the first row of numbers to see that gross income is really profit (total - cogs) where cogs is the cost of goods sold.  And that gross margin percentage is `profit/total`.

If profit is what we care about optimizing -- which it often is -- then gross income or gross margin percentage seems most relevant.  Let's focus on gross income (after all if we make a `$1000` on a 1% margin, that still sounds better than making `$10` where there's a 50% margin.

> We should be able to call a single function to do so.

```python
from item_lib import *

cols = ['Branch', 'City', 'Customer type', 'Gender', 'Product line']
target = 'gross income'
agg_name = 'percentage total profit'

results = build_and_print(supermarket_df, cols, target, agg = 'sum', agg_name = agg_name)
```

### Reviewing the data

One thing to note about the data above is that it's difficult to find anything too interesting in it. Why is that?  

The reason is because there's not much variation in the data.  Remember that at this stage, we are looking at categories to focus on.  But here, regardless of how we group our data -- whether by branch, customer type, gender, or product -- the data appears to be evenly split.
