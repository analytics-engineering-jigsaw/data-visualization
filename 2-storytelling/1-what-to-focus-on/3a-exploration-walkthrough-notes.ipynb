### 4. Digging Deeper

Once we have identified a couple of the key components that drive sales, we may want to look a bit closer at them.  

1. Selecting a subset and digging deeper

One way we can dig deeper is just by selecting a subset of the data and then digging deeper by calling our `build_and_print` on that subset.

For example, our Macbook sales account for 23% of revenue.  Let's just select Macbook sales and look more at those.

````python
macbook_df = df_with_order_date[df_with_order_date['Product'] == 'Macbook Pro Laptop']
````

And it will be nice to compare this against sales that are not macbooks.

````python
not_macbook_df = df_with_order_date[df_with_order_date['Product'] != 'Macbook Pro Laptop']
````

````python
df_with_order_date.columns
````

````python
target = 'Price Each'
agg_name = 'percent_revenue'
cols = ['City', 'order_Month', 'order_Week']

results = build_and_print(macbook_df, cols, target, agg = 'sum', agg_name = agg_name, y_range = [0, .25])
````

So here, we have looked at the distribution of just macbook sales.  Again, what we're interested is to see if there is anything interesting or surprising.  It turns out this breakdown looks pretty similar to non-macbook sales.

````python
target = 'Price Each'
agg_name = 'percent_revenue'
cols = ['City', 'order_Month', 'order_Week']

results = build_and_print(not_macbook_df, cols, target, agg = 'sum', agg_name = agg_name, y_range = [0, .25])
````

Perhaps one of the main differences is differences in the top order_Week of macbooks vs other products.  It may be worth digging deeper to see why macbook sales are higher in those weeks.

### 5. Findings and Next Steps

Ok, so this leads to the last step of our initial exploration.  Consolidating some of the findings, the hypotheses, and writing down some next steps.

For example, from the above we saw that:

1. Sales are concentrated in a few cities and across a few products.  
    * This is worth exploring further -- essentially, we want to learn more about why these products are dominating.  If it is just that these products are more expensive, is the company better at selling them than other competitors?  And are there particular stores or sales people who are particularly strong?
    
2. The breakdown by order week is surprising in that it did not appear sales were highly concentrated in the holiday season.  Are we underperforming competitors, and can we take advantage of better holiday sales?

3. We saw the Google Phone as an item that stands out in San Francisco.  Is there an opportunity to market more Google products there.  What accounts for the spike there, and is this something that can be replicated elsewhere?
